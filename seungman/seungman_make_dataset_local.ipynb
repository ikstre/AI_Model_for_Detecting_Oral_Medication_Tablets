# txt 변환용

```python
import os
import json
import glob
import shutil
import random

# =========================================================
# 🛠️ [경로 설정] 사용자님의 실제 폴더명으로 수정했습니다.
# =========================================================
base_path = r"C:\Users\home\Desktop\AI_study_1"

# 1. 아까 생성한 751개의 '완벽한 JSON' 폴더
json_root = os.path.join(base_path, "processed_json_new_fixed")

# 2. [수정됨] 751개의 '원본 이미지'가 들어있는 폴더 이름
# 사용자님이 어제 쓰셨던 'final_dataset_preprocessed_new_ONLY1.2'로 변경했습니다.
img_root = os.path.join(base_path, "final_dataset_preprocessed_new_ONLY1.2")

# 3. 결과물 저장 위치
output_root = os.path.join(base_path, "yolo_dataset_new_upload")
val_ratio = 0.2 

# =========================================================
# 🚀 실행 로직
# =========================================================

def convert_to_yolo_bbox(bbox, img_w, img_h):
    x, y, w, h = bbox
    cx, cy = x + (w / 2), y + (h / 2)
    return round(cx/img_w, 6), round(cy/img_h, 6), round(w/img_w, 6), round(h/img_h, 6)

def main():
    print(f"🚀 [751개 전용] 변환 작업을 시작합니다.")
    print(f"📂 이미지 폴더 수색: {img_root}")
    
    # 1. 이미지 주소록 생성 (대소문자 무시)
    img_lookup = {}
    
    # 경로가 실제로 존재하는지 먼저 확인
    if not os.path.exists(img_root):
        print(f"❌ [에러] 폴더를 찾을 수 없습니다: {img_root}")
        print("👉 바탕화면에서 해당 폴더의 이름을 다시 확인해 주세요.")
        return

    for root, dirs, files in os.walk(img_root):
        for f in files:
            if f.lower().endswith(('.jpg', '.jpeg', '.png')):
                full_path = os.path.join(root, f)
                img_lookup[f.lower()] = full_path
                img_lookup[os.path.splitext(f)[0].lower()] = full_path

    print(f"✅ 총 {len(img_lookup)//2}장의 이미지를 주소록에 등록했습니다.")
    
    if len(img_lookup) == 0:
        print(f"❌ [비상] '{img_root}' 폴더 안에 이미지가 한 장도 없습니다!")
        return

    # 폴더 초기화
    if os.path.exists(output_root): shutil.rmtree(output_root)
    for s in ['train', 'val']:
        os.makedirs(os.path.join(output_root, 'images', s), exist_ok=True)
        os.makedirs(os.path.join(output_root, 'labels', s), exist_ok=True)

    # 2. JSON 수집 및 클래스 매핑
    json_files = glob.glob(os.path.join(json_root, "**", "*.json"), recursive=True)
    print(f"📄 처리할 JSON: {len(json_files)}개")

    all_cats = {}
    for j in json_files:
        with open(j, 'r', encoding='utf-8') as f:
            d = json.load(f)
            for c in d.get('categories', []): all_cats[c['id']] = c['name']
    
    sorted_ids = sorted(all_cats.keys())
    id_to_idx = { drug_id: idx for idx, drug_id in enumerate(sorted_ids) }

    # 3. 변환 및 매칭
    random.shuffle(json_files)
    split_idx = int(len(json_files) * (1 - val_ratio))
    
    success, fail = 0, 0

    for i, jpath in enumerate(json_files):
        split = 'train' if i < split_idx else 'val'
        try:
            with open(jpath, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            orig_fname = data['images'][0]['file_name'].lower()
            orig_stem = os.path.splitext(orig_fname)[0].lower()
            
            src_path = img_lookup.get(orig_fname) or img_lookup.get(orig_stem)
            
            if not src_path:
                # [수정] 혹시 파일명에 'OLD_'가 붙어있는 경우 대응
                if orig_fname.startswith("old_"):
                    clean_name = orig_fname.replace("old_", "", 1)
                    src_path = img_lookup.get(clean_name) or img_lookup.get(os.path.splitext(clean_name)[0])

            if not src_path:
                print(f"❌ 매칭 실패: {orig_fname}")
                fail += 1
                continue

            # 복사 및 라벨 생성
            dst_fname = os.path.basename(src_path)
            shutil.copy2(src_path, os.path.join(output_root, 'images', split, dst_fname))
            
            txt_name = os.path.splitext(dst_fname)[0] + ".txt"
            with open(os.path.join(output_root, 'labels', split, txt_name), 'w', encoding='utf-8') as tf:
                for ann in data['annotations']:
                    c_idx = id_to_idx[ann['category_id']]
                    cx, cy, nw, nh = convert_to_yolo_bbox(ann['bbox'], data['images'][0]['width'], data['images'][0]['height'])
                    tf.write(f"{c_idx} {cx} {cy} {nw} {nh}\n")
            
            success += 1
        except Exception as e:
            print(f"🚫 에러: {e}")
            fail += 1

    # 4. data.yaml 생성
    names_list = [all_cats[uid] for uid in sorted_ids]
    yaml_txt = f"train: ../train/images\nval: ../val/images\nnc: {len(sorted_ids)}\nnames: {names_list}"
    with open(os.path.join(output_root, "data.yaml"), "w", encoding="utf-8") as f: f.write(yaml_txt)

    print(f"\n✅ 완료: {success}개 / 실패: {fail}개")
    if success > 0:
        print(f"📂 최종 데이터셋: {output_root}")

if __name__ == "__main__":
    main()
```

```python
import os

# =========================================================
# [설정] 경로를 정확하게 입력해주세요 (r"..." 사용하여 \ 문자 처리)
# =========================================================

# 1. 블랙리스트 (이 안에 있는 파일명들은 다 삭제 대상입니다)
BLACKLIST_DIR = r"C:\Users\home\Desktop\AI_Study\data\add_images\add2_images"

# 2. 삭제를 수행할 타겟 폴더 (이미지)
TARGET_IMG_DIR = r"C:\Users\home\Desktop\AI_Study_1\final_dataset_preprocessed_NEW_ONLY1.2"

# 3. 삭제를 수행할 타겟 폴더 (JSON)
TARGET_JSON_DIR = r"C:\Users\home\Desktop\AI_Study_1\processed_json_new_fixed"

# =========================================================

def get_filename_stems(folder_path):
    """
    폴더를 재귀적으로 탐색하여 확장자를 제외한 '파일명(stem)'들의 집합(Set)을 반환합니다.
    """
    stems = set()
    print(f"🔍 [블랙리스트 스캔 중] {folder_path} ...")
    
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # 확장자 제외한 파일명 추출 (예: dog.jpg -> dog)
            filename_no_ext = os.path.splitext(file)[0]
            stems.add(filename_no_ext)
            
    print(f"👉 총 {len(stems)}개의 삭제 대상(블랙리스트) 이름을 확보했습니다.")
    return stems

def delete_files_recursive(target_folder, blacklist_stems, file_type_msg):
    """
    타겟 폴더를 재귀적으로 돌면서 블랙리스트에 있는 이름이면 삭제합니다.
    """
    deleted_count = 0
    print(f"\n🚀 [{file_type_msg} 삭제 시작] {target_folder} 탐색 중...")

    for root, dirs, files in os.walk(target_folder):
        for file in files:
            filename_no_ext = os.path.splitext(file)[0]
            
            # 블랙리스트에 있는 이름이라면?
            if filename_no_ext in blacklist_stems:
                file_path = os.path.join(root, file)
                try:
                    os.remove(file_path)
                    print(f"❌ 삭제됨: {file_path}")
                    deleted_count += 1
                except Exception as e:
                    print(f"⚠️ 삭제 실패 ({file}): {e}")

    print(f"✅ [{file_type_msg}] 총 {deleted_count}개 파일 삭제 완료.")

def main():
    # 1. 경로가 실제로 존재하는지 확인
    if not os.path.exists(BLACKLIST_DIR):
        print("경로 에러: 블랙리스트 폴더가 없습니다.")
        return

    # 2. 블랙리스트 파일명(확장자 뺀 이름) 수집
    bad_stems = get_filename_stems(BLACKLIST_DIR)
    
    if not bad_stems:
        print("블랙리스트 폴더가 비어있습니다. 종료합니다.")
        return

    # 3. 타겟 이미지 폴더에서 삭제 수행
    if os.path.exists(TARGET_IMG_DIR):
        delete_files_recursive(TARGET_IMG_DIR, bad_stems, "이미지")
    else:
        print(f"경로 없음: {TARGET_IMG_DIR}")

    # 4. 타겟 JSON 폴더에서 삭제 수행
    if os.path.exists(TARGET_JSON_DIR):
        delete_files_recursive(TARGET_JSON_DIR, bad_stems, "JSON 라벨")
    else:
        print(f"경로 없음: {TARGET_JSON_DIR}")
        
    print("\n🎉 모든 작업이 끝났습니다.")

if __name__ == "__main__":
    main()
```

```python
import os

# =========================================================
# [설정] 경로 확인
# =========================================================
BLACKLIST_DIR = r"C:\Users\home\Desktop\AI_Study\data\add_images\add2_images"
TARGET_IMG_DIR = r"C:\Users\home\Desktop\AI_Study_1\final_dataset_preprocessed_NEW_ONLY1.2"
TARGET_JSON_DIR = r"C:\Users\home\Desktop\AI_Study_1\processed_json_new_fixed"
# =========================================================

def get_clean_stems(folder_path):
    stems = set()
    print(f"🔍 [블랙리스트] {folder_path} 로딩 중...")
    if not os.path.exists(folder_path): return stems

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # 소문자로 변환 + 공백 제거
            name = os.path.splitext(file)[0].lower().strip()
            stems.add(name)
            
    # 디버깅: 블랙리스트 파일명 예시 3개 출력
    print(f"👉 블랙리스트 예시: {list(stems)[:3]}")
    return stems

def delete_smart(target_folder, blacklist, label):
    print(f"\n🚀 [{label} 스마트 삭제 시작] {target_folder}...")
    
    if not os.path.exists(target_folder):
        print(f"❌ 경로 없음: {target_folder}")
        return

    deleted_count = 0
    
    for root, dirs, files in os.walk(target_folder):
        for file in files:
            # 1. 타겟 파일명 정제
            raw_name = os.path.splitext(file)[0].lower().strip()
            
            # 2. [핵심] 'new_' 제거하고 비교해보기
            # 예: 'new_k-001' -> 'k-001'로 바꿔서 블랙리스트에 있는지 확인
            name_without_new = raw_name.replace("new_", "")

            # 3. 비교 (그냥 같거나 OR new 떼고 같거나)
            if (raw_name in blacklist) or (name_without_new in blacklist):
                file_path = os.path.join(root, file)
                try:
                    os.remove(file_path)
                    deleted_count += 1
                    if deleted_count % 100 == 0:
                        print(f"   🗑️ {deleted_count}개 삭제 중... (예: {file})")
                except Exception as e:
                    print(f"⚠️ 에러: {e}")

    print(f"✅ [{label}] 총 {deleted_count}개 삭제 완료!")

# --- 실행 ---
bad_stems = get_clean_stems(BLACKLIST_DIR)

if bad_stems:
    delete_smart(TARGET_IMG_DIR, bad_stems, "이미지")
    delete_smart(TARGET_JSON_DIR, bad_stems, "JSON")
    print("\n🎉 작업 끝! 이제 'NEW_' 붙은 애들도 다 지워졌을 겁니다.")
else:
    print("블랙리스트가 비어있습니다.")
```

```python
import os
from tqdm import tqdm # 진행상황 보려고 추가함

# =========================================================
# [설정] 경로 확인
# =========================================================
BLACKLIST_DIR = r"C:\Users\home\Desktop\AI_Study\data\add_images\add2_images"
TARGET_IMG_DIR = r"C:\Users\home\Desktop\AI_Study_1\final_dataset_preprocessed_NEW_ONLY1.2"
TARGET_JSON_DIR = r"C:\Users\home\Desktop\AI_Study_1\processed_json_new_fixed"
# =========================================================

def get_clean_stems(folder_path):
    stems = set()
    print(f"🔍 [블랙리스트 로딩] {folder_path}")
    if not os.path.exists(folder_path): return stems

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # 확장자 떼고, 소문자로, 공백 제거
            name = os.path.splitext(file)[0].lower().strip()
            stems.add(name)
    
    # set을 list로 변환 (순서대로 대조하기 위해)
    return list(stems)

def delete_if_included(target_folder, blacklist_list, label):
    print(f"\n🚀 [{label} 강력 삭제 시작] {target_folder}...")
    
    if not os.path.exists(target_folder):
        print(f"❌ 경로 없음: {target_folder}")
        return

    # 1. 타겟 폴더의 모든 파일을 리스트업
    all_files = []
    for root, dirs, files in os.walk(target_folder):
        for file in files:
            all_files.append((root, file))

    print(f"   👉 검사할 파일 총 {len(all_files)}개 (시간이 좀 걸립니다)")

    deleted_count = 0
    
    # 2. 하나씩 꺼내서 검사 (tqdm으로 진행바 표시)
    for root, file in tqdm(all_files, desc=f"{label} 검사 중"):
        file_path = os.path.join(root, file)
        
        # 타겟 파일명 정제 (소문자)
        target_name = os.path.splitext(file)[0].lower()
        
        # 3. [핵심] 블랙리스트 단어가 타겟 이름 '안에' 들어있는지 확인
        # (예: target='new_abc_tail', black='abc') -> 포함되니까 삭제!
        is_bad = False
        for bad_name in blacklist_list:
            if bad_name in target_name:
                is_bad = True
                break # 범인 찾았으니 더 볼 필요 없음
        
        if is_bad:
            try:
                os.remove(file_path)
                deleted_count += 1
            except Exception as e:
                print(f"⚠️ 삭제 실패: {e}")

    print(f"✅ [{label}] 최종 삭제 완료: {deleted_count}개")

# --- 실행 ---
# 1. 블랙리스트 로딩
blacklist_list = get_clean_stems(BLACKLIST_DIR)
print(f"👉 블랙리스트 {len(blacklist_list)}개 장전 완료.")

if blacklist_list:
    # 2. 삭제 실행
    delete_if_included(TARGET_IMG_DIR, blacklist_list, "이미지")
    delete_if_included(TARGET_JSON_DIR, blacklist_list, "JSON")
    print("\n🎉 고생하셨습니다. 이번엔 진짜 다 지워졌을 겁니다.")
else:
    print("블랙리스트가 비어있습니다.")
```

# 데이터 선별(bbox오류 데이터 삭제)

```python
import os
from tqdm import tqdm

# ==============================================================
# ⚙️ [설정] 실행 모드 (중요!)
# ==============================================================
# True = 삭제는 안 하고 '삭제될 파일' 목록과 개수만 보여줍니다. (안전모드)
# False = 실제로 파일을 삭제합니다. (확인 후 변경하세요)
DRY_RUN = True  

# ==============================================================
# 📂 경로 설정
# ==============================================================
# 1. 생존자 명단 (기준이 되는 크롭 폴더)
CROP_DIR = r"C:\Users\home\Desktop\AI_study\processed_crops_old_ONLY1.2"

# 2. 정리 대상 (삭제 작업을 수행할 폴더들)
TARGET_DIRS = [
    r"C:\Users\home\Desktop\AI_study\data\train_annotations",
    r"C:\Users\home\Desktop\AI_study\data\train_images"
]

# ==============================================================
# 🚀 로직 시작
# ==============================================================

def get_whitelist_stems(crop_dir):
    """
    크롭 폴더를 뒤져서 '살려둘 파일의 이름(Stem)' 리스트를 만듭니다.
    조건: 파일명 앞의 'old_'를 제거하고 저장합니다.
    """
    whitelist = set()
    print(f"📋 기준 폴더 스캔 중: {crop_dir}")
    
    for root, dirs, files in os.walk(crop_dir):
        for file in files:
            # 이미지 파일인 경우만 체크 (필요시 확장자 추가 가능)
            if file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                
                # 1. 'old_' 접두어 제거 로직
                if file.startswith("old_"):
                    real_name = file[4:] # 앞의 4글자('old_') 자름
                else:
                    real_name = file # old_가 없으면 그대로 (혹시 몰라서)
                
                # 2. 확장자 제거하고 '이름'만 추출 (annotation과 매칭하기 위해)
                file_stem = os.path.splitext(real_name)[0]
                whitelist.add(file_stem)
                
    print(f"✅ 생존 리스트 확보 완료! (총 {len(whitelist)}개 파일 기준)")
    return whitelist

def clean_directory(target_dir, whitelist):
    """
    타겟 폴더를 돌면서 whitelist에 없는 파일을 삭제합니다.
    """
    print(f"\n🧹 정리 시작: {target_dir}")
    
    total_files = 0
    deleted_count = 0
    
    # 샅샅이 뒤지기 (os.walk)
    for root, dirs, files in os.walk(target_dir):
        for file in files:
            total_files += 1
            file_path = os.path.join(root, file)
            
            # 타겟 파일의 이름(Stem) 추출
            file_stem = os.path.splitext(file)[0]
            
            # 🚨 삭제 판단 로직
            # 내 이름이 생존 리스트에 없는가?
            if file_stem not in whitelist:
                if DRY_RUN:
                    # [모의 실행] 로그만 출력
                    # 너무 많으면 지저분하니까 10개만 예시로 출력하거나 주석 처리
                    # print(f"   [삭제 예정] {file}") 
                    pass
                else:
                    # [실제 실행] 삭제
                    try:
                        os.remove(file_path)
                    except Exception as e:
                        print(f"⚠️ 삭제 실패 ({file}): {e}")
                
                deleted_count += 1
            else:
                # 생존 (아무것도 안 함)
                pass

    return total_files, deleted_count

# ==============================================================
# 메인 실행부
# ==============================================================
if __name__ == "__main__":
    if DRY_RUN:
        print("🚧 [주의] 현재 '모의 실행(DRY_RUN)' 모드입니다. 파일이 실제로 삭제되지 않습니다.")
        print("   결과를 확인하고 코드를 수정하여 실제 삭제를 진행하세요.\n")
    else:
        print("🔥 [주의] '실제 삭제' 모드입니다. 복구할 수 없습니다!\n")

    # 1. 화이트리스트 만들기
    if not os.path.exists(CROP_DIR):
        print("❌ 기준 폴더(processed_crops)를 찾을 수 없습니다 경로를 확인하세요.")
        exit()
        
    valid_stems = get_whitelist_stems(CROP_DIR)

    # 2. 각 타겟 폴더 청소
    total_original_sum = 0
    total_deleted_sum = 0
    
    for target_dir in TARGET_DIRS:
        if os.path.exists(target_dir):
            orig, deleted = clean_directory(target_dir, valid_stems)
            
            remain = orig - deleted
            print(f"   👉 결과: {orig}개 중 {deleted}개 삭제 대상 -> {remain}개 남음")
            
            total_original_sum += orig
            total_deleted_sum += deleted
        else:
            print(f"\n⚠️ 타겟 폴더 없음: {target_dir}")

    # 3. 최종 리포트
    print("\n" + "="*50)
    print(f"📊 최종 집계 ({'모의 실행' if DRY_RUN else '실제 삭제 완료'})")
    print(f"   - 전체 파일 수 : {total_original_sum}")
    print(f"   - 삭제된 파일 수: {total_deleted_sum}")
    print(f"   - 남은 파일 수  : {total_original_sum - total_deleted_sum}")
    print("="*50)
    
    if DRY_RUN:
        print("\n💡 실제 삭제를 하려면 코드 맨 위 'DRY_RUN = True'를 'False'로 바꾸세요.")
```

```python
import os
import json
import shutil
import re
from tqdm import tqdm
from pathlib import Path
import random

# ==============================================================
# 📂 [경로 설정] 본인 환경에 맞게 수정하세요!
# ==============================================================
# 1. 생존자 명단 (크롭 이미지가 있는 폴더)
CROP_DIR = r"C:\Users\home\Desktop\AI_study\processed_crops_new_ONLY1.2"

# 2. 원본 데이터 (JSON 폴더)
RAW_DATA_DIR = r"C:\Users\home\Desktop\AI_study\data\add_annotations"

# 3. 원본 이미지 (이미지 폴더)
RAW_IMG_DIR = r"C:\Users\home\Desktop\AI_study\data\add_images"

# 4. 결과 저장 폴더 (새로 생성됨)
OUTPUT_DIR = r"C:\Users\home\Desktop\AI_study\YOLO_Dataset_Final_Strict"


# ==============================================================
# 🔑 [족보] 팀원들과 맞춘 ID 매핑 (global_category_index.json 기반)
# ==============================================================
# K-코드의 숫자 부분(String) -> YOLO Class Index(Int)
# 이 맵핑표는 절대 수정하지 마세요! (팀 공통 기준)
ID_TO_INDEX = {
    "1900": 0, "2483": 1, "3351": 2, "3483": 3, "3544": 4, "3743": 5, "3832": 6, "4543": 7,
    "12081": 8, "12247": 9, "12778": 10, "13395": 11, "13900": 12, "16232": 13, "16262": 14,
    "16548": 15, "16551": 16, "16688": 17, "18147": 18, "18357": 19, "19232": 20, "19552": 21,
    "19607": 22, "19861": 23, "20014": 24, "20238": 25, "20877": 26, "21325": 27, "21771": 28,
    "22074": 29, "22347": 30, "22362": 31, "24850": 32, "25367": 33, "25438": 34, "25469": 35,
    "27733": 36, "27777": 37, "27926": 38, "27993": 39, "28763": 40, "29345": 41, "29451": 42,
    "29667": 43, "30308": 44, "31863": 45, "31885": 46, "32310": 47, "33009": 48, "33208": 49,
    "33880": 50, "34597": 51, "35206": 52, "36637": 53, "38162": 54, "41768": 55
}

# yolo_classes.txt 순서대로 이름 리스트 생성 (data.yaml 생성용)
# 인덱스 0번부터 55번까지 순서대로 이름이 들어가야 합니다.
CLASS_NAMES = [
    "보령부스파정 5mg", "뮤테란캡슐 100mg", "일양하이트린정 2mg", "기넥신에프정(은행엽엑스)(수출용)",
    "무코스타정(레바미피드)(비매품)", "알드린정", "뉴로메드정(옥시라세탐)", "에어탈정(아세클로페낙)",
    "리렉스펜정 300mg/PTP", "아빌리파이정 10mg", "다보타민큐정 10mg/병", "써스펜8시간이알서방정 650mg",
    "에빅사정(메만틴염산염)(비매품)", "리피토정 20mg", "크레스토정 20mg", "가바토파정 100mg",
    "동아가바펜틴정 800mg", "오마코연질캡슐(오메가-3-산에틸에스테르90)", "리리카캡슐 150mg", "종근당글리아티린연질캡슐(콜린알포세레이트)",
    "콜리네이트연질캡슐 400mg", "트루비타정 60mg/병", "스토가정 10mg", "노바스크정 5mg",
    "마도파정", "플라빅스정 75mg", "엑스포지정 5/160mg", "아토르바정 10mg",
    "라비에트정 20mg", "리피로우정 20mg", "자누비아정 50mg", "맥시부펜이알정 300mg",
    "놀텍정 10mg", "자누메트정 50/850mg", "큐시드정 31.5mg/PTP", "아모잘탄정 5/100mg",
    "트윈스타정 40/5mg", "카나브정 60mg", "울트라셋이알서방정", "졸로푸트정 100mg",
    "트라젠타정(리나글립틴)", "비모보정 500/20mg", "레일라정", "리바로정 4mg",
    "트라젠타듀오정 2.5/850mg", "아질렉트정(라사길린메실산염)", "자누메트엑스알서방정 100/1000mg", "글리아타민연질캡슐",
    "신바로정", "에스원엠프정 20mg", "글리틴정(콜린알포세레이트)", "제미메트서방정 50/1000mg",
    "아토젯정 10/40mg", "로수젯정10/5밀리그램", "로수바미브정 10/20mg", "카발린캡슐 25mg"
]

# ==============================================================
# 🛠️ 로직 시작
# ==============================================================

def get_yolo_class_id(k_code):
    """
    K-Code(예: K-001900)를 받아서 YOLO Class ID(예: 0)를 반환
    """
    try:
        # 1. 숫자만 추출 (001900)
        # 2. 정수로 변환 후 다시 문자열로 (1900) -> 앞의 0 제거 효과
        # 3. 맵핑 테이블에서 조회
        numeric_part = str(int(k_code.replace("K-", "")))
        return ID_TO_INDEX.get(numeric_part)
    except:
        return None

def parse_expected_ids(filename_stem):
    """
    파일명에서 '원래 있어야 할 ID들'을 추출
    예: K-001900-016548... -> {'K-001900', 'K-016548'...}
    """
    id_part = filename_stem.split('_')[0]
    expected_ids = set(re.findall(r"(K-\d{6})", id_part))
    return expected_ids

def get_survivor_map(crop_dir):
    """
    크롭 폴더 스캔 -> { '원본파일스템': {'살아남은ID1', ...} }
    """
    survivor_map = {}
    print("📋 [1단계] 생존자 명단(크롭 파일) 스캔 중...")
    
    files = list(Path(crop_dir).rglob("*.*"))
    for f in tqdm(files):
        if f.suffix.lower() in ['.jpg', '.png', '.jpeg']:
            name_body = f.stem
            
            # NEW_ 접두어 제거
            if name_body.upper().startswith("NEW_"):
                name_body = name_body[4:]
            
            # _K-xxxx 접미어 분리
            if "_K-" in name_body:
                original_stem, drug_id = name_body.rsplit("_", 1)
                if original_stem not in survivor_map:
                    survivor_map[original_stem] = set()
                survivor_map[original_stem].add(drug_id)
                
    print(f"✅ 총 {len(survivor_map)}개의 원본 파일이 생존자 명단에 등록되었습니다.")
    return survivor_map

def convert_to_yolo_bbox(img_w, img_h, bbox):
    # JSON: x, y, w, h -> YOLO: cx, cy, w, h
    x, y, w, h = bbox
    cx = (x + w / 2) / img_w
    cy = (y + h / 2) / img_h
    nw = w / img_w
    nh = h / img_h
    return cx, cy, nw, nh

def main():
    # 1. 생존자 맵 생성
    survivor_map = get_survivor_map(CROP_DIR)
    
    if not survivor_map:
        print("❌ 크롭 폴더에서 유효한 파일을 찾지 못했습니다. 경로를 확인해주세요!")
        return

    # 2. 폴더 생성 및 yaml 작성
    print("\n📋 [2단계] 학습용 폴더 및 data.yaml 생성 중...")
    for split in ['train', 'val']:
        os.makedirs(os.path.join(OUTPUT_DIR, 'images', split), exist_ok=True)
        os.makedirs(os.path.join(OUTPUT_DIR, 'labels', split), exist_ok=True)

    # data.yaml 생성 (팀원 클래스 순서 완벽 적용)
    yaml_path = os.path.join(OUTPUT_DIR, 'data.yaml')
    with open(yaml_path, 'w', encoding='utf-8') as f:
        f.write(f"path: {OUTPUT_DIR}\n")
        f.write("train: images/train\n")
        f.write("val: images/val\n")
        f.write(f"nc: {len(CLASS_NAMES)}\n")
        f.write("names:\n")
        for i, name in enumerate(CLASS_NAMES):
            f.write(f"  {i}: {name}\n")
    
    # 3. JSON 매칭 및 파일 생성 (Strict Mode)
    print("\n🚀 [3단계] 원본 데이터 검증 및 생성 시작 (엄격 모드)...")
    json_files = list(Path(RAW_DATA_DIR).rglob("*.json"))
    
    # JSON 그룹화 (파일명 기준)
    image_data_buffer = {}
    for json_path in json_files:
        try:
            with open(json_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
                if 'images' in data and data['images']:
                    file_name = data['images'][0]['file_name']
                    stem = os.path.splitext(file_name)[0]
                    if stem not in image_data_buffer:
                        image_data_buffer[stem] = []
                    image_data_buffer[stem].append(data)
        except: continue

    pass_count = 0
    drop_count = 0
    
    for stem, json_list in tqdm(image_data_buffer.items()):
        # A. 검증: 원래 있어야 할 약들(파일명) vs 실제 생존한 약들(크롭폴더)
        expected_ids = parse_expected_ids(stem)
        survived_ids = survivor_map.get(stem, set())
        
        # 🚨 엄격한 필터링: 하나라도 부족하면 탈락!
        if not expected_ids or not expected_ids.issubset(survived_ids):
            drop_count += 1
            continue
            
        # B. 합격 -> 라벨링 진행
        # 이미지 파일 찾기
        first_img_name = json_list[0]['images'][0]['file_name']
        found_imgs = list(Path(RAW_IMG_DIR).rglob(first_img_name))
        
        if not found_imgs:
            # 대소문자 이슈 등으로 못 찾으면 스킵 (필요시 로직 추가 가능)
            continue
        src_img_path = found_imgs[0]
        
        # 라벨 데이터 생성
        label_lines = []
        for data in json_list:
            drug_id = data['images'][0].get('drug_N') # 예: K-001900
            
            # ID 매핑 (K-001900 -> 0)
            class_id = get_yolo_class_id(drug_id)
            
            if class_id is not None:
                if 'annotations' in data and data['annotations']:
                    bbox = data['annotations'][0]['bbox']
                    w = data['images'][0]['width']
                    h = data['images'][0]['height']
                    
                    yolo_bbox = convert_to_yolo_bbox(w, h, bbox)
                    
                    # YOLO 포맷: class_id cx cy w h
                    line = f"{class_id} {yolo_bbox[0]:.6f} {yolo_bbox[1]:.6f} {yolo_bbox[2]:.6f} {yolo_bbox[3]:.6f}"
                    label_lines.append(line)
        
        if label_lines:
            pass_count += 1
            # Train/Val 랜덤 분할 (8:2)
            split = 'train' if random.random() < 0.8 else 'val'
            
            # TXT 저장
            txt_path = os.path.join(OUTPUT_DIR, 'labels', split, stem + ".txt")
            with open(txt_path, 'w') as f:
                f.write('\n'.join(label_lines))
            
            # 이미지 복사
            dst_img_path = os.path.join(OUTPUT_DIR, 'images', split, first_img_name)
            shutil.copy(src_img_path, dst_img_path)

    print("\n" + "="*50)
    print(f"📊 최종 결과")
    print(f"   - ✅ 생성된 데이터셋: {pass_count}장 (완벽한 이미지)")
    print(f"   - 🗑️ 폐기된 이미지: {drop_count}장 (불량 포함)")
    print(f"   - 저장 경로: {OUTPUT_DIR}")
    print("="*50)

if __name__ == "__main__":
    main()
```

```python
import os
import json
import shutil
import glob
from pathlib import Path
from tqdm import tqdm
import random
import cv2

# ==============================================================
# 📂 [경로 설정]
# ==============================================================
CROP_DIR = r"C:\Users\home\Desktop\AI_study\processed_crops_NEW_ONLY1.2"
RAW_DATA_DIR = r"C:\Users\home\Desktop\AI_Study\data\add_annotations" 
RAW_IMG_DIR = r"C:\Users\home\Desktop\AI_Study\data\add_images" 
OUTPUT_DIR = r"C:\Users\home\Desktop\AI_study\YOLO_Dataset_New_Smart_Strict" 

# ==============================================================
# 🔑 [족보] 56종 ID 매핑
# ==============================================================
ID_TO_INDEX = {
    "1900": 0, "2483": 1, "3351": 2, "3483": 3, "3544": 4, "3743": 5, "3832": 6, "4543": 7,
    "12081": 8, "12247": 9, "12778": 10, "13395": 11, "13900": 12, "16232": 13, "16262": 14,
    "16548": 15, "16551": 16, "16688": 17, "18147": 18, "18357": 19, "19232": 20, "19552": 21,
    "19607": 22, "19861": 23, "20014": 24, "20238": 25, "20877": 26, "21325": 27, "21771": 28,
    "22074": 29, "22347": 30, "22362": 31, "24850": 32, "25367": 33, "25438": 34, "25469": 35,
    "27733": 36, "27777": 37, "27926": 38, "27993": 39, "28763": 40, "29345": 41, "29451": 42,
    "29667": 43, "30308": 44, "31863": 45, "31885": 46, "32310": 47, "33009": 48, "33208": 49,
    "33880": 50, "34597": 51, "35206": 52, "36637": 53, "38162": 54, "41768": 55
}

CLASS_NAMES = [
    "보령부스파정 5mg", "뮤테란캡슐 100mg", "일양하이트린정 2mg", "기넥신에프정(은행엽엑스)(수출용)",
    "무코스타정(레바미피드)(비매품)", "알드린정", "뉴로메드정(옥시라세탐)", "에어탈정(아세클로페낙)",
    "리렉스펜정 300mg/PTP", "아빌리파이정 10mg", "다보타민큐정 10mg/병", "써스펜8시간이알서방정 650mg",
    "에빅사정(메만틴염산염)(비매품)", "리피토정 20mg", "크레스토정 20mg", "가바토파정 100mg",
    "동아가바펜틴정 800mg", "오마코연질캡슐(오메가-3-산에틸에스테르90)", "리리카캡슐 150mg", "종근당글리아티린연질캡슐(콜린알포세레이트)",
    "콜리네이트연질캡슐 400mg", "트루비타정 60mg/병", "스토가정 10mg", "노바스크정 5mg",
    "마도파정", "플라빅스정 75mg", "엑스포지정 5/160mg", "아토르바정 10mg",
    "라비에트정 20mg", "리피로우정 20mg", "자누비아정 50mg", "맥시부펜이알정 300mg",
    "놀텍정 10mg", "자누메트정 50/850mg", "큐시드정 31.5mg/PTP", "아모잘탄정 5/100mg",
    "트윈스타정 40/5mg", "카나브정 60mg", "울트라셋이알서방정", "졸로푸트정 100mg",
    "트라젠타정(리나글립틴)", "비모보정 500/20mg", "레일라정", "리바로정 4mg",
    "트라젠타듀오정 2.5/850mg", "아질렉트정(라사길린메실산염)", "자누메트엑스알서방정 100/1000mg", "글리아타민연질캡슐",
    "신바로정", "에스원엠프정 20mg", "글리틴정(콜린알포세레이트)", "제미메트서방정 50/1000mg",
    "아토젯정 10/40mg", "로수젯정10/5밀리그램", "로수바미브정 10/20mg", "카발린캡슐 25mg"
]

# ==============================================================
# 🛠️ 함수 정의
# ==============================================================

def parse_survivor_filename(filename):
    """
    크롭 파일명에서 '원본 파일명'과 '해당 약의 ID'를 추출
    예: OLD_K-001900-..._200_K-001900.jpg 
    -> 원본: K-001900-..._200, 약ID: 1900
    """
    stem = Path(filename).stem
    
    # 1. 접두어 제거 (OLD_, NEW_ 등)
    # 대소문자 구분 없이 처리하기 위해 체크
    if stem.upper().startswith("OLD_"):
        stem = stem[4:]
    elif stem.upper().startswith("NEW_"):
        stem = stem[4:]
        
    # 2. 접미어 (_K-xxxx) 분리
    # 파일명 중간에 _K-가 있을 수 있으니 오른쪽 끝(rsplit) 기준으로 분리
    if "_K-" in stem:
        original_stem, drug_part = stem.rsplit("_K-", 1)
        drug_id = drug_part  # 예: 001900
        
        # 숫자만 남기기 (혹시 모를 문자 제거)
        drug_id_clean = "".join(filter(str.isdigit, drug_id))
        return original_stem, drug_id_clean
    
    return None, None

def scan_raw_data_map(raw_data_dir):
    """
    RAW_DATA_DIR을 뒤져서 { '파일명': { '약ID': 'JSON경로' } } 형태의 맵을 생성
    * 핵심: 폴더명(K-xxxx)을 보고 약 ID를 판별함
    """
    print("📋 [준비] 원본 라벨(JSON) 위치 파악 중...")
    json_map = {} # Key: filename_stem, Value: { drug_id: full_path }
    
    # rglob으로 모든 json 탐색
    all_jsons = list(Path(raw_data_dir).rglob("*.json"))
    
    for p in tqdm(all_jsons):
        stem = p.stem
        # 부모 폴더 이름 확인 (예: K-001900)
        parent_folder = p.parent.name
        
        if parent_folder.startswith("K-"):
            drug_id = parent_folder.replace("K-", "")
            # 0 제거하여 매칭 확률 높임 (001900 -> 1900) -> 문자열 통일을 위해 그대로 둠 or 조심
            # 사용자 데이터는 "001900" 6자리 형태 유지하는게 안전함.
            
            if stem not in json_map:
                json_map[stem] = {}
            
            # 같은 파일명이라도 폴더(약ID)별로 경로 저장
            json_map[stem][drug_id] = str(p)
            
    return json_map

def convert_bbox(w, h, bbox):
    # x, y, bw, bh
    x_min, y_min, width, height = bbox
    
    cx = (x_min + width / 2) / w
    cy = (y_min + height / 2) / h
    nw = width / w
    nh = height / h
    
    # 좌표 정규화 (0~1) 클램핑
    cx = max(0, min(1, cx))
    cy = max(0, min(1, cy))
    nw = max(0, min(1, nw))
    nh = max(0, min(1, nh))
    
    return cx, cy, nw, nh

# ==============================================================
# 🚀 메인 실행
# ==============================================================

def main():
    # 1. 생존자 명단 작성 (크롭 파일 스캔)
    survivor_files = list(Path(CROP_DIR).rglob("*.*"))
    survivors = {} # Key: 원본파일명(stem), Value: set(생존한 약ID들)
    
    print(f"📋 [1단계] 생존자(크롭) {len(survivor_files)}개 스캔 중...")
    
    for f in tqdm(survivor_files):
        if f.suffix.lower() not in ['.jpg', '.png', '.jpeg']:
            continue
            
        org_stem, drug_id = parse_survivor_filename(f.name)
        if org_stem and drug_id:
            # 약ID 앞의 0을 제거해서 매핑표(ID_TO_INDEX)와 맞춤 (예: 001900 -> 1900)
            # 단, 폴더명 매칭을 위해 원본 형태도 고려해야 함.
            # 여기서는 int로 변환했다가 다시 str로 하면 "1900"이 됨.
            clean_id = str(int(drug_id)) 
            
            if org_stem not in survivors:
                survivors[org_stem] = set()
            survivors[org_stem].add(clean_id) # Set에 추가 (중복 방지)

    print(f"✅ 총 {len(survivors)}개의 '원본 이미지'가 복구 대상입니다.")

    # 2. 원본 JSON 위치 맵핑 (속도 향상 및 정확성 위해)
    json_path_map = scan_raw_data_map(RAW_DATA_DIR)

    # 3. 폴더 생성
    for split in ['train', 'val']:
        os.makedirs(os.path.join(OUTPUT_DIR, 'images', split), exist_ok=True)
        os.makedirs(os.path.join(OUTPUT_DIR, 'labels', split), exist_ok=True)

    # data.yaml 생성
    with open(os.path.join(OUTPUT_DIR, 'data.yaml'), 'w') as f:
        f.write(f"path: {OUTPUT_DIR}\n")
        f.write("train: images/train\n")
        f.write("val: images/val\n")
        f.write(f"nc: {len(CLASS_NAMES)}\n")
        f.write("names:\n")
        for i, n in enumerate(CLASS_NAMES):
            f.write(f"  {i}: {n}\n")

    # 4. 데이터셋 구축 (이미지 복사 + 라벨 생성)
    print("\n🚀 [3단계] 데이터셋 구축 시작 (조건: 크롭이 존재하면 무조건 살림)")
    
    count_saved = 0
    count_no_json = 0
    count_no_img = 0

    # 생존자 리스트에 있는 원본 파일들만 순회
    for org_stem, survivor_drug_ids in tqdm(survivors.items()):
        
        # (1) 원본 이미지 찾기
        # 확장자가 png일지 jpg일지 모르니 검색
        img_candidates = list(Path(RAW_IMG_DIR).rglob(f"{org_stem}.*"))
        valid_img_path = None
        for p in img_candidates:
            if p.suffix.lower() in ['.jpg', '.png', '.jpeg']:
                valid_img_path = p
                break
        
        if not valid_img_path:
            count_no_img += 1
            continue
            
        # (2) 라벨 데이터 생성
        yolo_lines = []
        
        # 이 이미지가 가진 '살아남은 약'들을 하나씩 처리
        for drug_id in survivor_drug_ids:
            # 56종 타겟인지 확인
            if drug_id not in ID_TO_INDEX:
                continue
            
            class_idx = ID_TO_INDEX[drug_id]
            
            # JSON 파일 찾기 (JSON 맵 활용)
            # json_map의 키는 drug_id가 "001900"일 수도 "1900"일 수도 있음. 
            # 사용자 설명상 폴더명은 "K-001900" -> drug_id 추출시 "001900"
            # 하지만 ID_TO_INDEX는 "1900".
            # 매칭을 위해 두 가지 경우 다 시도.
            
            target_json_path = None
            if org_stem in json_path_map:
                # Case A: 맵에 1900으로 저장된 경우
                if drug_id in json_path_map[org_stem]:
                    target_json_path = json_path_map[org_stem][drug_id]
                # Case B: 맵에 001900으로 저장된 경우 (0 padding)
                else:
                    padded_id = drug_id.zfill(6)
                    if padded_id in json_path_map[org_stem]:
                        target_json_path = json_path_map[org_stem][padded_id]
            
            if not target_json_path:
                # 크롭은 있는데 원본 JSON을 못 찾음 (경로 문제 등)
                continue
                
            # JSON 파싱 및 좌표 추출
            try:
                with open(target_json_path, 'r', encoding='utf-8') as f:
                    data = json.load(f)
                    
                    # 이미지 사이즈 (정규화용)
                    img_w = data['images'][0]['width']
                    img_h = data['images'][0]['height']
                    
                    # BBox
                    if 'annotations' in data and len(data['annotations']) > 0:
                        bbox = data['annotations'][0]['bbox'] # [x, y, w, h]
                        
                        # YOLO 변환 (스마트 필터링 OFF -> 계산만 수행)
                        cx, cy, nw, nh = convert_bbox(img_w, img_h, bbox)
                        
                        # 라벨 한 줄 추가
                        yolo_lines.append(f"{class_idx} {cx:.6f} {cy:.6f} {nw:.6f} {nh:.6f}")
            except Exception as e:
                print(f"Error parsing {target_json_path}: {e}")
                continue

        # (3) 저장 결정
        # 유효한 라벨이 하나라도 있으면 저장 (즉, 생존자가 하나라도 있으면)
        if yolo_lines:
            split = 'train' if random.random() < 0.8 else 'val'
            
            # 라벨 파일 저장
            txt_save_path = os.path.join(OUTPUT_DIR, 'labels', split, org_stem + ".txt")
            with open(txt_save_path, 'w') as f:
                f.write("\n".join(yolo_lines))
                
            # 이미지 파일 복사
            img_save_path = os.path.join(OUTPUT_DIR, 'images', split, valid_img_path.name)
            shutil.copy(valid_img_path, img_save_path)
            
            count_saved += 1
        else:
            # 크롭 파일은 있었으나, 56종이 아니거나 JSON 매칭 실패로 라벨이 0개인 경우
            count_no_json += 1

    print("\n" + "="*50)
    print("📊 최종 완료 보고")
    print(f"  - ✅ 생성된 데이터셋 이미지 수: {count_saved}장")
    print(f"  - ❌ 원본 이미지를 못 찾음: {count_no_img}장")
    print(f"  - ⚠️ JSON 매칭 실패 또는 타겟(56종) 아님: {count_no_json}장")
    print(f"  - 저장 경로: {OUTPUT_DIR}")
    print("="*50)

if __name__ == "__main__":
    main()
```

# 라벨 데이터 확인용 코드

```python
import os
import glob
from pathlib import Path
from collections import Counter

# ==============================================================
# 📂 [경로 확인] 방금 만든 데이터셋 경로를 넣어주세요
# ==============================================================
DATASET_DIR = r"C:\Users\home\Desktop\AI_study\YOLO_Dataset_New_Smart_Strict_7787(2)"

def verify_dataset():
    print("🔎 데이터셋 정밀 검증 시작...")
    
    label_files = list(Path(DATASET_DIR).rglob("*.txt"))
    
    if not label_files:
        print("❌ 라벨 파일이 하나도 없습니다! 경로를 확인하세요.")
        return

    total_files = len(label_files)
    total_bboxes = 0
    bbox_count_distribution = [] # 각 파일에 몇 개의 박스가 있는지 기록

    for p in label_files:
        with open(p, 'r') as f:
            lines = f.readlines()
            # 공백 줄 제거하고 실제 데이터 줄만 카운트
            valid_lines = [line for line in lines if line.strip()]
            count = len(valid_lines)
            
            total_bboxes += count
            bbox_count_distribution.append(count)

    # 분포 계산
    dist_counter = Counter(bbox_count_distribution)
    sorted_dist = sorted(dist_counter.items())

    print("\n" + "="*40)
    print(f"📊 [검증 결과 리포트]")
    print("="*40)
    print(f"1️⃣ 총 라벨 파일(이미지) 수: {total_files}장")
    print(f"2️⃣ 총 발견된 BBox(알약) 수: {total_bboxes}개")
    print("-" * 40)
    print("3️⃣ 파일당 알약 개수 분포:")
    for count, num_files in sorted_dist:
        print(f"   - 알약 {count}개 있는 파일: {num_files}장")
    print("-" * 40)
    
    # 평가
    avg = total_bboxes / total_files if total_files else 0
    print(f"💡 평균: 이미지 1장당 약 {avg:.2f}개의 알약이 라벨링됨.")
    
    # 아까 크롭 개수(약 22,821개)와 비교
    print(f"\n[판정]")
    if 22000 <= total_bboxes <= 23500:
        print("✅ 성공! 크롭된 생존자 수(약 22,821개)와 거의 일치합니다.")
    else:
        print(f"⚠️ 경고! 예상된 크롭 수(22,821)와 차이가 큽니다. ({22821 - total_bboxes}개 차이)")

if __name__ == "__main__":
    verify_dataset()
```
